```python
# helper to load data from PNG image files
#import imageio
import scipy.misc
# glob helps select multiple files using patterns
import glob
import csv
```

```python
import numpy
# library for plotting arrays
import matplotlib.pyplot
# ensure the plots are inside this notebook, not an external window
%matplotlib inline
```

```python

#for image_file_name in glob.glob('straight_neutral_png/train/*.png'):
out_file = open('face_train.csv', 'w', newline='')
out_csv = csv.writer(out_file)

for image_file_name in glob.glob('resource/face_resource/train/*.png'):
    print ("loading ... ", image_file_name)

    #label = 1, if "sunglass" in file_name, label = 0 : if "sunglass" is not in file_name
    if(image_file_name.find("straight") >= 0):
        label = 0
        pass
    elif(image_file_name.find("left") >=0):
        label = 1
        pass
    elif(image_file_name.find("right") >=0):
        label = 2
        pass
    # load image data from png files into an array
    img_array = scipy.misc.imread(image_file_name, flatten=True, mode='I')
    # reshape from 32x30 to list of 960 values, invert values
    img_data = 255 - img_array.reshape(960)
    
    # matplotlib.pyplot.imshow(img_data.reshape(30,32), cmap='Greys', interpolation='None')
        
    # append label and image data  to test data set
    record = numpy.append(label,img_data)
    irecord = [ int(f) for f in record] 
    #print(record)
    out_csv.writerow(irecord)
    pass

out_file.close()

##---------------test data 

out_file = open('face_test.csv', 'w', newline='')
out_csv = csv.writer(out_file)

for image_file_name in glob.glob('resource/face_resource/test/*.png'):
    print ("loading ... ", image_file_name)

    #label = 1, if "sunglass" in file_name, label = 0 : if "sunglass" is not in file_name
    if(image_file_name.find("straight") >= 0):
        label = 0
        pass
    elif(image_file_name.find("left") >=0):
        label = 1
        pass
    elif(image_file_name.find("right") >=0):
        label = 2
        pass
    # load image data from png files into an array
    img_array = scipy.misc.imread(image_file_name, flatten=True)
    # reshape from 32x30 to list of 960 values, invert values
    img_data = 255 - img_array.reshape(960)
    
    # matplotlib.pyplot.imshow(img_data.reshape(30,32), cmap='Greys', interpolation='None')
        
    # append label and image data  to test data set
    record = numpy.append(label,img_data)
    irecord = [ int(f) for f in record] 
    #print(record)
    out_csv.writerow(irecord)
    pass

out_file.close()
```
